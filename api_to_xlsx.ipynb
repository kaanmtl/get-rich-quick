```python
import requests
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
```

```python
import csv

stock_codes = []
with open('bist100kod.csv', newline='') as inputfile:
    for row in csv.reader(inputfile):
        stock_codes.append(row[0])
```

```python
df=pd.DataFrame()
#stock_codes=['TUKAS','GARAN','YKBNK']
for code in stock_codes:
    try:
        date = requests.get(f"https://www.oyakyatirim.com.tr/Equity/GetHistoricalEquityData?mode=3&code={code}")
        date = np.array(eval(date.text))[:,0]
        vals = requests.get(f"https://www.oyakyatirim.com.tr/Equity/GetHistoricalEquityData?mode=3&code={code}")
        vals = np.array(eval(vals.text))[:,1]
    except Exception as e:
        print(str(e))
    
    try:
        df['DATE']=date
        df[code]=vals
    except Exception as e:
        print(str(e))
df.to_excel(r'C:\Users\Kaan\Desktop\data\output.xlsx', index = False)
```
