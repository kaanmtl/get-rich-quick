```python
import requests
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
```

```python
stock_codes=['TUKAS','GARAN','YKBNK']
df=pd.DataFrame()
for code in stock_codes:
    
    date = requests.get(f"https://www.oyakyatirim.com.tr/Equity/GetHistoricalEquityData?mode=2&code={code}")
    date = np.array(eval(date.text))[:,0]
    
    vals = requests.get(f"https://www.oyakyatirim.com.tr/Equity/GetHistoricalEquityData?mode=2&code={code}")
    vals = np.array(eval(vals.text))[:,1]
    
    df['DATE']=date
    df[code]=vals
df.to_excel(r'C:\Users\Kaan\Desktop\data\output.xlsx', index = False)
```
